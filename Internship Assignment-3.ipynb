```python
#filling the missing values with 0 so that we get equal no of row counts overall
df1.fillna(0, inplace= True)
```

```python
#finding total number of males and females
df1["Sex"].value_counts()
```

```python
#BAR CHART using for representiong the number of males and females
df1["Sex"].value_counts().plot(kind = "bar",color= "maroon")
plt.title("Number of males and females")
plt.xlabel("Sex")
plt.ylabel("Numbers")
plt.show()
```

The above data states that females are more in HR field as compared to Males.

```python
#segregating no of Females and Males as per department and counting the values
grouped = df1.groupby(['Department', 'Sex'])['Sex'].count().unstack()
grouped = grouped.fillna(0)
grouped
```

```python
#HORIZONTAL BAR CHART with department wise gender distribution in a form of stack
grouped.plot(kind="barh", stacked = True)
plt.title("Distribution of Gender by Department")
plt.ylabel("Department")
plt.xlabel("Numbers")
plt.legend(["Female", "Male"])
plt.show()
```

In the above graph we can see both males and females have larger count in production department and both are very few in Executive office department

```python
#PIE CHART for the mariral description
marital_counts = df1['MaritalDesc'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(marital_counts, labels=marital_counts.index, autopct='%1.1f%%', startangle=45)
plt.axis('equal') 
plt.show()
```

The above graph showcases the percentage of people who are either married, divorced , seperated,
widowed or single

```python
#CATPLOT between marital status id and salary
sns.catplot(data=df1, x='MaritalStatusID', y='Salary', hue='GenderID', kind='box')
plt.show()
```

The above diagram is the representation of the categorical plot which shows how the salaries are affected with respect to the marital status. Here category 1 and 2 has maximum salary in the range between 5000 to 100000
